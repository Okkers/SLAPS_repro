```python
import matplotlib.pyplot as plt
import numpy as np

slaps = {"FP": 69.96, "MLP": 69.60, "MLP-D": 71.02}
slaps_2s = {"FP": 70.16, "MLP": 65.45, "MLP-D": 69.46}

labels = list(slaps.keys())
slaps_values = list(slaps.values())
slaps_2s_values = list(slaps_2s.values())

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(8, 6))
rects1 = ax.bar(x - width/2, slaps_values, width, label='SLAPS')
rects2 = ax.bar(x + width/2, slaps_2s_values, width, label='SLAPS_2S')

ax.set_ylabel('Accuracy')
ax.set_title('Figure 3 - reproduced')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend(loc='lower right')

# Add value labels on top of the bars
ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

# Expand y-axis slightly to accommodate the labels on the 100 values
ax.set_ylim(0, 110)

plt.tight_layout()

plt.savefig('figure_3.png')
```
